```python
import sklearn as sk
import pandas as pd
from pathlib import Path
from numpy.random import seed
path = Path('../dataset_new/Global Economy Indicators New Cleaned.csv')
ds_new = pd.read_csv(path)
ds_new_y = ds_new
ds_new_x = ds_new.drop(columns=[' IMF based exchange rate '])
```

## Splitting The data into train and test

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
seed(42)
x_train,x_test,y_train,y_test = train_test_split(ds_new_x,ds_new_y,test_size=0.8,train_size=0.2,random_state=42,)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
precision = r2_score(y_test,y_pred)
precision
```
